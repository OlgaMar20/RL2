#### **Machine Learning** 
<p>  ML stands for "Machine learning" in English, and refers to a branch of artificial intelligence that focuses on the development of algorithms and techniques that allow computers to learn through experience and data, rather than being explicitly programmed to perform a task.

In other words, machine learning is based on the idea that machines can improve their performance on a task as they receive more data and feedback, and use this information to adjust their models and algorithms.

Machine learning is applied in a wide variety of fields, including image and voice recognition, natural language processing, fraud and anomaly detection, financial risk prediction and business process optimization.</p>

#### **Types of Automatic Learning**
<p>Machine learning can be divided into two main categories: supervised learning and unsupervised learning.

<Strong>Supervised learning </strong> is one in which a set of labeled data is provided to the learning algorithm, i.e. data that already has a classification or label assigned to it. The goal of the algorithm is to learn how to classify new data or predict output values based on the input data. For example, a supervised learning algorithm can be trained to classify emails as spam or not spam, or to predict the price of a house based on its characteristics.

On the other hand, unsupervised <strong>learning </strong> is one in which no labels are provided for the data. The goal of the algorithm is to discover patterns and structures in the data without any specific guidance. For example, an unsupervised learning algorithm can be used to group customers of a store into different categories based on their shopping behavior, or to find patterns in large data sets without labels.

We can affirm that, while supervised learning focuses on learning to predict output values from labeled data, unsupervised learning focuses on discovering patterns and structures in the data without any specific guidance.</p>

##### **Types of ML Algorithms**
There are several types of algorithms in machine learning (ML), each of them with its own specific technique and purpose. Here are some examples:

1. Regression: It is used to predict continuous values, such as the price of a house or a person's salary.

2. Classification: It is used to predict discrete values, such as whether an email is spam or not, or whether an image contains a cat or a dog.

3. Clustering: It is used to group similar data into groups or clusters, such as customers with similar purchasing behaviors.

4. Decision trees: They are used to represent decisions and actions in a tree-like format, where each node represents a decision and each branch represents an action.

5. Neural networks: They are used to mimic the functioning of the human brain and are used for complex classification and prediction tasks.

6. Deep learning: A subcategory of neural networks that uses multiple layers of processing to learn complex features of data.

7. Reinforcement learning: It is used to train models that can make optimal decisions and actions in a specific environment, based on the feedback and rewards received.

These are just a few examples of the many types of ML algorithms that exist, and the <strong>choice of the right algorithm will depend on the specific task that you want to solve </strong>.

##### **Single or Multiple Linear Regression**

<p>Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. In its simplest form, it is a linear function that is fitted to a data set to predict values of the dependent variable based on the values of the independent variables.

In linear regression, we start from the hypothesis that the relationship between the dependent variable and the independent variables is linear, which means that the dependent variable can be expressed as a linear combination of the independent variables.

The objective of linear regression is to find the values of the parameters of the linear function that best fit the data. This is achieved by minimizing the sum of the squared errors between the actual values of the dependent variable and the values predicted by the model.

Linear regression is used in a wide variety of fields, including economics, psychology, biology, engineering and data science. It is a simple but powerful technique that can provide valuable information about the relationship between variables.  </p>

### **Simple Linear Regression with Python**
The *statsmodel* package helps to elaborate the linear regression

```python
# Importamos librerias
import pandas as pd
import numpy as np
```

```python
# Leemos dataset csv
data = pd.read_csv("Sales_200_days.csv")
```

```python
# Explorar el dataset
data.head()
```

```python
data.shape
```

```python
# Importamos statsmodels para obtener la regresión lineal 
import statsmodels.formula.api as smf
```

##### Simple Linear Regression Sales ~ TV

```python
# Relacionamos la variable independiente TV con la variable dependiente Ventas (Sales)
lm = smf.ols(formula="Sales~TV", data = data).fit()
```

```python
lm.params

# Nota el valor de predicción sería Sales = 7.032594 + 0.047537 * TV 
```

```python
# Determinar el valor del coeficiente de predictibilidad
lm.pvalues
```

```python
print(lm.summary())
```

```python
# Error de predictibilidad
sales_pred = lm.predict(pd.DataFrame(data["TV"]))
sales_pred
```

```python
data['PrediccionTV']= 7.032594 + 0.047537 * data['TV'] 
```

```python
data.head()
```

```python
data['ErrorTV'] = data['Sales'] - data['PrediccionTV']
```

```python
data.head()
```

```python
#Graficar la regresión lineal
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
data.plot(kind = "scatter", x = "TV", y="Sales")
# Recta de regresión lineal
plt.plot(pd.DataFrame(data["TV"]), sales_pred, c = "red", linewidth= 3)
```

### **Personal reflection about the simple regression exercise**


Linear regression is a powerful and widely used technique in statistics and data science that allows us to model the relationship between a dependent variable and one or more independent variables. This technique is based on the assumption that there is a linear relationship between variables, which means that changes in an independent variable are linearly related to changes in the dependent variable.

Simple linear regression is suitable when we have only one independent variable that we want to use to predict the dependent variable. However, in many cases, the relationship between the dependent variable and the independent variables may be more complex and cannot be completely captured by a single independent variable. This is where multiple linear regression comes into play.

Multiple linear regression allows us to model the relationship between a dependent variable and two or more independent variables. By including multiple independent variables in the model, we can capture and take into account the combined influence of these variables on the dependent variable. This allows us to make more accurate predictions and capture more complex relationships between variables.

However, it is important to remember that multiple linear regression also has its limitations. For example, it assumes a linear relationship between variables and may not adequately capture non-linear relationships. In addition, the inclusion of too many independent variables can lead to overfitting of the model, which can result in unreliable predictions in new data.

In summary, both simple linear regression and multiple linear regression are useful tools for modeling and understanding the relationships between variables in data. By understanding the strengths and limitations of each approach, we can effectively use them to make predictions and make informed decisions in a variety of applications.
